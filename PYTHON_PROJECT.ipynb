# Task
Calculate Recency, Frequency, and Monetary (RFM) metrics for each customer using the `final_df` DataFrame. Group the data by `CustomerID` to find the latest `TransactionDate` for Recency, the total count of `TransactionDate` for Frequency, and the sum of `TransactionAmount` for Monetary. Then, determine a reference date (e.g., the day after the latest transaction in the dataset) and calculate Recency in days for each customer. Store these RFM values in a new DataFrame called `df_rfm`, and finally, display the first few rows of `df_rfm` to verify the calculations.

```python
rfm_df = final_df.groupby('CustomerID').agg(
    Recency=('TransactionDate', 'max'),
    Frequency=('TransactionDate', 'count'),
    Monetary=('TransactionAmount', 'sum')
).reset_index()

rfm_df.head()
```

**Reasoning**:
Now that Recency, Frequency, and Monetary values have been calculated and stored in `rfm_df`, the next step is to calculate the Recency in days. This requires determining a reference date, which will be the day after the latest transaction date across all customers in `rfm_df`, and then subtracting each customer's `Recency` (latest transaction date) from this reference date.



```python
reference_date = rfm_df['Recency'].max() + pd.Timedelta(days=1)
rfm_df['Recency_days'] = (reference_date - rfm_df['Recency']).dt.days

# Display the updated rfm_df with Recency_days
rfm_df.head()
```

## Create RFM DataFrame

### Subtask:
Store the calculated Recency, Frequency, and Monetary values in a new DataFrame named `df_rfm`.

# Task
Calculate R, F, and M scores for customers in `df_rfm` using quintiles, where lower `Recency_days` receives a higher `R_Score` (5 for the lowest quintile, 1 for the highest) and higher `Frequency` and `Monetary` values receive higher `F_Score` and `M_Score` respectively (5 for the highest quintile, 1 for the lowest). Finally, display the `df_rfm` DataFrame with the new `R_Score`, `F_Score`, and `M_Score` columns.

```python
df_rfm['R_Score'] = pd.qcut(df_rfm['Recency_days'], q=5, labels=False, duplicates='drop')
df_rfm['R_Score'] = df_rfm['R_Score'].max() - df_rfm['R_Score'] + 1
df_rfm.head()
```

**Reasoning**:
To calculate the Frequency Score, I will use `pd.qcut` to divide the `Frequency` into 5 quantiles. Since higher `Frequency` should correspond to higher scores, I'll generate integer bins using `labels=False` and then shift them to assign scores from 1 to 5. The `duplicates='drop'` argument will handle any non-unique quantile edges.



```python
df_rfm['F_Score'] = pd.qcut(df_rfm['Frequency'], q=5, labels=False, duplicates='drop') + 1
df_rfm.head()
```

**Reasoning**:
To calculate the Monetary Score, I will use `pd.qcut` to divide the `Monetary` into 5 quantiles. Since higher `Monetary` values should correspond to higher scores, I'll generate integer bins using `labels=False` and then shift them to assign scores from 1 to 5. The `duplicates='drop'` argument will handle any non-unique quantile edges.



```python
df_rfm['M_Score'] = pd.qcut(df_rfm['Monetary'], q=5, labels=False, duplicates='drop') + 1
df_rfm.head()
```

## Combined Percentage Revenue Contribution by R, F, and M Scores

### Subtask:
Create a combined bar graph to visualize and compare the percentage of total revenue contributed by each `R_Score`, `F_Score`, and `M_Score` level.

**Reasoning**:
To combine and compare the revenue contribution across R, F, and M scores, I will first prepare the data by concatenating the individual `revenue_by_r_score`, `revenue_by_f_score`, and `revenue_by_m_score` DataFrames. I will add a 'Score_Type' column to differentiate between Recency, Frequency, and Monetary scores. Then, I will use `seaborn.catplot` to create a grouped bar chart, allowing for a direct comparison of the percentage revenue contribution for each score level across the three RFM dimensions. This provides a compact and effective way to observe their relative impacts.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure total_revenue is defined
if 'total_revenue' not in locals():
    total_revenue = df_rfm['Monetary'].sum()

# Prepare data for R_Score
revenue_by_r_score = df_rfm.groupby('R_Score')['Monetary'].sum().reset_index()
revenue_by_r_score['Percentage_Contribution'] = (revenue_by_r_score['Monetary'] / total_revenue) * 100
revenue_by_r_score['Score_Type'] = 'Recency'
revenue_by_r_score = revenue_by_r_score.rename(columns={'R_Score': 'Score'})

# Prepare data for F_Score
revenue_by_f_score = df_rfm.groupby('F_Score')['Monetary'].sum().reset_index()
revenue_by_f_score['Percentage_Contribution'] = (revenue_by_f_score['Monetary'] / total_revenue) * 100
revenue_by_f_score['Score_Type'] = 'Frequency'
revenue_by_f_score = revenue_by_f_score.rename(columns={'F_Score': 'Score'})

# Prepare data for M_Score
revenue_by_m_score = df_rfm.groupby('M_Score')['Monetary'].sum().reset_index()
revenue_by_m_score['Percentage_Contribution'] = (revenue_by_m_score['Monetary'] / total_revenue) * 100
revenue_by_m_score['Score_Type'] = 'Monetary'
revenue_by_m_score = revenue_by_m_score.rename(columns={'M_Score': 'Score'})

# Concatenate all three dataframes
combined_revenue_scores = pd.concat([revenue_by_r_score, revenue_by_f_score, revenue_by_m_score])

# Create a grouped bar chart
plt.figure(figsize=(18, 10)) # Increased figure size
g = sns.catplot(
    data=combined_revenue_scores,
    x='Score',
    y='Percentage_Contribution',
    hue='Score_Type',
    kind='bar',
    palette='Dark2', # Changed palette for more distinct colors
    height=7, # Slightly increased height
    aspect=2, # Increased aspect ratio
    legend_out=True # Ensure legend is outside if desired
)

# Add data labels on top of the bars
for ax in g.axes.flat:
    for p in ax.patches:
        if p.get_height() > 0:
            ax.annotate(f'{p.get_height():.1f}%',
                        (p.get_x() + p.get_width() / 2., p.get_height()),
                        ha='center', va='center',
                        xytext=(0, 9),
                        textcoords='offset points',
                        fontsize=9,
                        color='black')

g.set_axis_labels("Score (1-5)", "Percentage of Total Revenue (%)") # Improved y-axis label
g.set_titles("Revenue Contribution by {col_name} Score")
plt.suptitle('Combined Percentage Revenue Contribution by RFM Scores', y=1.02, fontsize=20, weight='bold') # Improved and bolded title
plt.tight_layout(rect=[0, 0.03, 1, 0.98]) # Adjusted rect for suptitle
plt.show()
```

## Assign Segment Labels

### Subtask:
Use business rules to define segment labels for selected score combinations, such as "555" for "Champion" and "111" for "Lost", and apply these to create a new column called `RFM_Segment_Label`.

**Reasoning**:
To assign segment labels, I will define a function that takes the `RFM_Segment` string and returns a descriptive label based on predefined business rules. I'll then apply this function to the `RFM_Segment` column to create a new `RFM_Segment_Label` column. This approach allows for flexible and extensive mapping of RFM segments to various customer types.

```python
def assign_rfm_label(rfm_segment):
    r = int(rfm_segment[0])
    f = int(rfm_segment[1])
    m = int(rfm_segment[2])

    if r >= 4 and f >= 4 and m >= 4:
        return 'Champion'
    elif r >= 4 and f >= 3 and m >= 3:
        return 'Loyal Customers'
    elif r >= 3 and f >= 3 and m >= 3:
        return 'Potential Loyalist'
    elif r >= 3 and f <= 2 and m <= 2:
        return 'New Customers'
    elif r <= 2 and f >= 3 and m >= 3:
        return 'At Risk'
    elif r <= 2 and f <= 2 and m <= 2:
        return 'Lost'
    else:
        return 'Other'

df_rfm['RFM_Segment_Label'] = df_rfm['RFM_Segment'].apply(assign_rfm_label)
df_rfm.head()
```

**Reasoning**:
To perform a Pareto analysis on revenue, I will first sort the customers by their monetary value in descending order. Then, I will calculate the cumulative percentage of monetary value and the cumulative percentage of customers. Finally, I will plot these two cumulative percentages to visualize the 80/20 rule.



```python
df_pareto = df_rfm.sort_values(by='Monetary', ascending=False).reset_index(drop=True)
df_pareto['Cumulative_Monetary'] = df_pareto['Monetary'].cumsum()
df_pareto['Cumulative_Monetary_Percentage'] = 100 * df_pareto['Cumulative_Monetary'] / df_pareto['Monetary'].sum()
df_pareto['Cumulative_Customer_Percentage'] = 100 * (df_pareto.index + 1) / len(df_pareto)

plt.figure(figsize=(12, 6))
sns.lineplot(x='Cumulative_Customer_Percentage', y='Cumulative_Monetary_Percentage', data=df_pareto)
plt.axvline(x=20, color='r', linestyle='--', label='20% Customers')
plt.axhline(y=80, color='g', linestyle='--', label='80% Revenue')
plt.title('Pareto Analysis (Cumulative Revenue vs. Cumulative Customers)', fontsize=16)
plt.xlabel('Cumulative Percentage of Customers', fontsize=12)
plt.ylabel('Cumulative Percentage of Revenue', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```
